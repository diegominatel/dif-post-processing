## Charts

```python
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import warnings

warnings.filterwarnings("ignore") 

df_plot = report_plot()
```

```python
order = ['ad', 'ada', 'lr', 'mlp', 'rf', 'xgb']
plt.rcParams.update({'font.size': 18})

g = sns.FacetGrid(df_plot, col="Config", height=4, aspect=.5, col_order=order)

palette = 'Set2'
colors = sns.color_palette(palette)

def barplot_no_labels(data, **kwargs):
    sns.barplot(data=data, **kwargs)
    ax = plt.gca()
    ax.set_xticklabels([])  
    ax.set_xlabel('')     
    
g.map_dataframe(barplot_no_labels, x="Postp", y="MCPM",  palette=palette, ci=None)

for ax in g.axes.flat:
    ax.axhline(y=1.268, color='red', linestyle='--', linewidth=1)

handles = [Patch(color=colors[0], label='WPP'), Patch(color=colors[1], label='DIF-PP'),
           Patch(color=colors[2], label='CEOP'), Patch(color=colors[3], label='ROBC')]

g.add_legend(title="")
g.fig.legend(handles=handles, loc='lower center', bbox_to_anchor=(0.5, -0.05), ncol=9)

plt.subplots_adjust(top=0.85)

for ax in g.axes.flat:
    ax.set_xlabel('')
    ax.set_ylabel('MCPM')
    
new_titles = ['AD', 'ADA', 'LR', 'MLP', 'RF', 'XGB']
for ax, new_title in zip(g.axes.flat, new_titles):
    ax.set_title(new_title)

sns.despine(trim=False, offset=10)

plt.savefig("multicriteria_by_algorithm.pdf", format='pdf', bbox_inches='tight')
plt.show()
```

## Analysis of DIF-PP Parameters

```python
df = raw_report()
df = df[df['postp'] == 'dif-postp']
```

```python
# atualiza parmetro T
df.loc[df['method'] == '2PL_N-16_P-0.05', 'T'] = "(0.05; 0.95)"
df.loc[df['method'] == '2PL_N-16_P-0.1', 'T'] = "(0.10; 0.90)"
df.loc[df['method'] == '2PL_N-16_P-0.2', 'T'] = "(0.20; 0.80)"
df.loc[df['method'] == '2PL_N-16_P-0.4', 'T'] = "(0.40; 0.60)"
df.loc[df['method'] == '2PL_N-16_P-0.45', 'T'] = "(0.45; 0.55)"
df.loc[df['method'] == '2PL_N-32_P-0.05', 'T'] = "(0.05; 0.95)"
df.loc[df['method'] == '2PL_N-32_P-0.1', 'T'] = "(0.10; 0.90)"
df.loc[df['method'] == '2PL_N-32_P-0.2', 'T'] = "(0.20; 0.80)"
df.loc[df['method'] == '2PL_N-32_P-0.4', 'T'] = "(0.40; 0.60)"
df.loc[df['method'] == '2PL_N-32_P-0.45', 'T'] = "(0.45; 0.55)"
# atualiza parmetro k
df.loc[df['method'] == '2PL_N-16_P-0.05', 'k'] = "16"
df.loc[df['method'] == '2PL_N-16_P-0.1', 'k'] = "16"
df.loc[df['method'] == '2PL_N-16_P-0.2', 'k'] = "16"
df.loc[df['method'] == '2PL_N-16_P-0.4', 'k'] = "16"
df.loc[df['method'] == '2PL_N-16_P-0.45', 'k'] = "16"
df.loc[df['method'] == '2PL_N-32_P-0.05', 'k'] = "32"
df.loc[df['method'] == '2PL_N-32_P-0.1', 'k'] = "32"
df.loc[df['method'] == '2PL_N-32_P-0.2', 'k'] = "32"
df.loc[df['method'] == '2PL_N-32_P-0.4', 'k'] = "32"
df.loc[df['method'] == '2PL_N-32_P-0.45', 'k'] = "32"
```

```python
df = df.groupby(by=['T', 'k']).mean()
```

```python
df = df[['multicriteria', 'f1_macro']]
df = df.reset_index()
```

```python
plt.rcParams.update({'font.size': 14})

custom_palette = {'16': 'palevioletred', '32': 'darkturquoise'}

# Plotando com seaborn lineplot
plt.figure(figsize=(8, 6))
g = sns.lineplot(
    data=df, x='T', y='multicriteria', hue='k', style='k', linewidth=2,
    markers=['s','o'], dashes=True, palette=custom_palette
)

for line in g.get_lines():
    line.set_markersize(8)

leg = plt.legend(title='k', loc=3)
leg.get_frame().set_facecolor('none')
leg.get_frame().set_linewidth(0.0)

g.set_xlabel('R')
g.set_ylabel('MCPM')


sns.despine(trim=False, offset=10)

plt.savefig("parametro_mcpm.pdf", format='pdf', bbox_inches='tight')
plt.show()
```
